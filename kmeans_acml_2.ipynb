```python
import numpy as np
import random


#  squared Euclidean distance (__d__) between a data point x and a centroid c
def __distance__(x, c):
    s = 0.0
    for x_i, c_i in zip(x, c):
        s += pow((x_i-c_i), 2)
    return s

# calculates sum of squared errors for set of clusters and their corresponding centroids.
def __loss__(clusters, centres):
    s = 0.0
    for c in clusters:
        si = 0.0
        for x in clusters[c]:
            si += __distance__(x, centres[c])
        s += si
    return s


def __kmeans__(D, centres):
    k = 3
    iter = 2

    for epoch in range(iter):

        # randomness
        random.shuffle(D)

        # store in dict structure
        clusters = {i: [] for i in range(len(centres))}

        # print(clusters)

        # assign x to cluster j using minimal distance
        for x in D:
            distances = []
            for c in centres:
                distances.append(__distance__(x, c))
            min_d = min(distances)
            min_d_idx = distances.index(min_d)
            clusters[min_d_idx].append(x)
        
        # print(clusters)

        # compute loss
        loss = __loss__(clusters, centres)
        print("%.4f" % loss)

        # move cluster centre j to mean (mu) of the points in cluster
        for i, c in enumerate(centres):
            mu = np.zeros_like(c) # ([a,b])
            for x in clusters[i]:
                for j in range(len(x)):
                    mu[j] += x[j]
            mu = mu/len(clusters[i])
            centres[i] = mu


if __name__ == '__main__':

    inputs = []
    for i in range(6):
        i = input()
        if i == '-1':
            break
        inputs.append(float(i))

    # stoeing the centres
    centres = []
    for i in range(len(inputs)-1):
        if i % 2 == 0:
            centres.append([inputs[i], inputs[i+1]])

    # centres = [
    #     [0.45, 0.55],
    #     [0.70, 0.71],
    #     [0.11, 0.67]
    # ]

    D = [
        [0.22, 0.33],
        [0.45, 0.76],
        [0.73, 0.39],
        [0.25, 0.35],
        [0.51, 0.69],
        [0.69, 0.42],
        [0.41, 0.49],
        [0.15, 0.29],
        [0.81, 0.32],
        [0.50, 0.88],
        [0.23, 0.31],
        [0.77, 0.30],
        [0.56, 0.75],
        [0.11, 0.38],
        [0.81, 0.33],
        [0.59, 0.77],
        [0.10, 0.89],
        [0.55, 0.09],
        [0.75, 0.35],
        [0.44, 0.55]
    ]

    __kmeans__(D, centres)
```
